### Part 2: Fuel used increases linearly with number of steps to move

- Moving 1 costs 1
- Moving 2 costs 3 (1 + 2)
- moving 3 costs 6 (1 + 2 + 3)
- moving 4 costs 10 (1 + 2 + 3 + 4)

Triangle numbers? 

suppose steps = 4
range(1,steps+1) will yield 1,2,3,4
simply add these values to the running fuel count

```python
def fuel_used_steps(steps:int):
    '''
    calculates how much fuel is used if increases linearly 
    with number of steps
    '''
    fuel_count = 0
    for i in range(1,steps+1):
        fuel_count += i
    return fuel_count

fuel_used_steps(4)
```
