```python
def replace_number_with_letter(number: int, board: list):
    '''
    When input number matches number on board, 
    it is replaced with a letter
    '''
    new_board = []
    for row in board:
        new_row = [i if number!=i else 'x' for i in row]
        new_board.append(new_row)
    return new_board
    
```

```python
def check_for_win(board: list):
    '''
    Checks rows and columns for winning streak of x values
    '''
    
    for row in board:
        if row.count('x') == 5:
            return True

    # Check verticals
    for i in range(len(board[0])):
        x_counter = 0
        for row in board:
            if row[i] == 'x':
                x_counter += 1
        if x_counter == 5:
            return True
    
    # if both loops complete, no win yet
    return False
```

```python
def create_board(list_of_rows: list):
    
    board = []
    for row in list_of_rows:
        split_row = row.split(' ')
        split_row = list(filter(lambda a:a!='',split_row))
        split_row = [int(i) for i in split_row]
        board.append(split_row)
    return board
```
