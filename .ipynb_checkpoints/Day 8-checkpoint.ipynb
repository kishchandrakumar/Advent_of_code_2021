#### Ordering matters

```python
a = 'abcde'
b = 'abc'

b in a
```

```python
a = 'abcde'
b = ''.join(sorted('cab'))

b in a
```

```python
for i,digits in enumerate(inp):
    inp[i] = ''.join(sorted(digits))
    
inp
```

```python
number_tracker = {}

for i in inp: 
    if len(i) == 2:
        number_tracker[1] = sorted(i)
    elif len(i) == 3:
        number_tracker[7] = sorted(i)
    elif len(i) == 4:
        number_tracker[4] = sorted(i)
    elif len(i) == 7:
        number_tracker[8] = sorted(i)
```

```python
number_tracker
```

```python
for i in inp: 
    letters = sorted(i)
    if len(i) == 6: # 0, 6 or 9
        if all(j in letters for j in number_tracker[4]):
            number_tracker[9] = letters
        elif all(j in letters for j in number_tracker[7]):
            number_tracker[0] = letters
        else:
            number_tracker[6] = letters
            
    elif len(i) == 5: #2, 3 or 5
        if all(j in letters for j in number_tracker[7]):
            number_tracker[3] = letters
        elif len(set(letters).intersection(number_tracker[4])) == 2:
            number_tracker[2] = letters
        else:
            number_tracker[5] = letters

number_tracker
```

```python
values = 'cdfeb fcadb cdfeb cdbaf'.split(' ')
values
```

```python
out_number = ''
for num in values: 
    # check which key in number_tracker has exactly the letters in num
    digit = sorted(num)
    index = list(number_tracker.values()).index(digit)
    key = list(number_tracker.keys())[index]
    
    # append the string of this key to out_number
    out_number += str(key)
    
print(out_number)
```

## Actual input for part 2

```python
with open('day8_input.txt','r') as f:
    read_file = f.read().split('\n')
# outputs = read_file.split('\n')


digits_0_to_9 = []
digits_to_convert = []
for i in read_file:
    split = i.split(' | ')
    digits_0_to_9.append(split[0])
    digits_to_convert.append(split[1])
    
```

```python
digits_0_to_9[0:10]
```

```python
digits_to_convert[0:10]
```

```python
number_trackers = []


for row in digits_0_to_9:
    inp = row.split(' ')
    for i,digits in enumerate(inp):
        inp[i] = ''.join(sorted(digits))
        
    number_tracker = {}

    for i in inp: 
        if len(i) == 2:
            number_tracker[1] = sorted(i)
        elif len(i) == 3:
            number_tracker[7] = sorted(i)
        elif len(i) == 4:
            number_tracker[4] = sorted(i)
        elif len(i) == 7:
            number_tracker[8] = sorted(i)
            
    for i in inp: 
        letters = sorted(i)
        if len(i) == 6: # 0, 6 or 9
            if all(j in letters for j in number_tracker[4]):
                number_tracker[9] = letters
            elif all(j in letters for j in number_tracker[7]):
                number_tracker[0] = letters
            else:
                number_tracker[6] = letters

        elif len(i) == 5: #2, 3 or 5
            if all(j in letters for j in number_tracker[7]):
                number_tracker[3] = letters
            elif len(set(letters).intersection(number_tracker[4])) == 2:
                number_tracker[2] = letters
            else:
                number_tracker[5] = letters
            
    number_trackers.append(number_tracker)
```

```python
number_trackers
```

```python
out_numbers = []
for i in range(len(number_trackers)):
    values = digits_to_convert[i].split(' ')
    out_number = ''
    for num in values: 
        # check which key in number_tracker has exactly the letters in num
        digit = sorted(num)
        index = list(number_trackers[i].values()).index(digit)
        key = list(number_trackers[i].keys())[index]

        # append the string of this key to out_number
        out_number += str(key)
    
    out_numbers.append(int(out_number))
    
```
