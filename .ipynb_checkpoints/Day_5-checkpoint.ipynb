```python
with open('day5_input.txt', 'r') as f:
    read_f = f.read()
```

```python
print(read_f)
```

```python
lines =[i.split(' -> ') for i in  read_f.split('\n')]
lines
```

```python
source = []
destination = []
for coords in lines:
    source.append(coords[0])
    destination.append(coords[1])
```

```python
source
```

```python
source_coords = []
for coord in source:
    x_and_y = coord.split(',')
    source_coords.append(dict(x = int(x_and_y[0]), y = int(x_and_y[1])))

target_coords = []
for coord in destination:
    x_and_y = coord.split(',')
    target_coords.append(dict(x = int(x_and_y[0]), y = int(x_and_y[1])))
```

```python
source_coords
```

### Consider horizontal lines only first:


```python
import numpy as np
```

```python
empty_grid = np.zeros(shape=(1000,1000), dtype=int)
```

```python
horizontal_indices = []

for i in range(len(source_coords)):
    if source_coords[i]['y'] == target_coords[i]['y']:
        horizontal_indices.append(i)
        
len(horizontal_indices)
```

```python
for index in horizontal_indices:
    max_x = max(source_coords[index]['x'],target_coords[index]['x'])
    min_x = min(source_coords[index]['x'],target_coords[index]['x'])
    y = source_coords[index]['y']
    for i in range(min_x, max_x+1):
        empty_grid[i][y] += 1
```

```python
empty_grid[empty_grid > 1].shape
```

```python
vertical_indices = []

for i in range(len(source_coords)):
    if source_coords[i]['x'] == target_coords[i]['x']:
        vertical_indices.append(i)

len(vertical_indices)
```

```python
for index in vertical_indices:
    max_y = max(source_coords[index]['y'],target_coords[index]['y'])
    min_y = min(source_coords[index]['y'],target_coords[index]['y'])
    x = source_coords[index]['x']
    for i in range(min_y, max_y+1):
        empty_grid[x][i] += 1
```

```python
empty_grid[empty_grid > 1].shape
```

## Horizontal 

```python
diagonal_indices = []

for i in range(len(source_coords)):
    if abs(source_coords[i]['y']-target_coords[i]['y']) == abs(source_coords[i]['x']-target_coords[i]['x']):
        diagonal_indices.append(i)

len(diagonal_indices)
```

```python
for index in diagonal_indices:
    # Choose the minimum x value as source 
    if source_coords[index]['x'] < target_coords[index]['x']:
        source_x = source_coords[index]['x']
        source_y = source_coords[index]['y']
        target_y = target_coords[index]['y']
        y_diff = target_y - source_y
        if y_diff >= 0: # up-trend case
            for i in range(y_diff+1):
                empty_grid[source_x+i][source_y+i] += 1
        if y_diff < 0: # down-trend case
            for i in range(y_diff+1):
                empty_grid[source_x+i][source_y-i] += 1
            
        
    elif source_coords[index]['x'] > target_coords[index]['x']:
        source_x = target_coords[index]['x']
        source_y = target_coords[index]['y']
        target_y = source_coords[index]['y']
        y_diff = target_y - source_y
        if y_diff >= 0: # up-trend case
            for i in range(y_diff+1):
                empty_grid[source_x+i][source_y+i] += 1
        if y_diff < 0: # down-trend case
            for i in range(y_diff+1):
                empty_grid[source_x+i][source_y-i] += 1
```

```python
empty_grid[empty_grid > 1].shape
```
